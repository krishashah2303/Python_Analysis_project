```python
#Counting occurence of each skill
#storing it in key value form

from collections import Counter
input =  df['skills']
cnt = Counter( input )
print( cnt.items() )
```

```python
#Garphical representation of skills and their occurence in the dataset

plt.figure(figsize=(20,20))
names = list(cnt.keys())
values = list(cnt.values())
plt.barh(range(len(cnt)), values,tick_label=names, edgecolor="grey")
plt.xlabel("Values", fontsize=20)
plt.ylabel("skills", fontsize=20)
plt.title('skills and their occurence', fontsize=20)
plt.legend()
plt.show()
```

```python
#Counting occurence of each hobbie

from collections import Counter
input =  df['hobbies']
c = Counter( input )
print( c.items() )
```

```python
#Garphical representation of hobbies and their occurence in the dataset

plt.figure(figsize=(15,12))
names = list(c.keys())
values = list(c.values())
plt.barh(range(len(c)), values,tick_label=names,color='red', edgecolor="blue")
plt.xlabel("Values", fontsize=20)
plt.ylabel("Hobbies", fontsize=20)
plt.title('Hobbies and their occurence', fontsize=20)
plt.legend()
plt.show()
```

```python
#Counting occurence of each carrer

from collections import Counter
input =  df['career']
career = Counter( input )

print( career.items() )
```

```python
#Garphical representation of career and their occurence in the dataset

plt.figure(figsize=(8,5))
names = list(career.keys())
values = list(career.values())
plt.barh(range(len(career)), values,tick_label=names,color='green', edgecolor="yellow")
plt.xlabel("Values", fontsize=12)
plt.ylabel("Career", fontsize=12)
plt.title('Career and their occurence')
plt.legend()
plt.show()
```
